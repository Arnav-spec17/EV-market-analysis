```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
raw_data=pd.read_csv('IEA-EV-data.csv')
raw_data.iloc[3201:3336,:]
r_data=raw_data.iloc[3201:3334,:]
```

```python
r_data
```

```python
india_data=raw_data[raw_data['region']=='India']

fig,ax=plt.subplots()
ax.bar(india_data['mode'], india_data['value'])
ax.set_title('Electric Vehicles in India')
ax.set_xlabel('Mode')
ax.set_ylabel('Number of Electric Vehicles')
plt.show()
```

```python
raw_data1=pd.read_csv('RS_Session_254_AU_2384.C.i.csv')
raw_data1
```

```python
transport_data=raw_data1.loc[0:14, ['Vehicle Class', 'Total Registered on VAHAN 4.0']]
non_transport_data=raw_data1.loc[15:26, ['Vehicle Class', 'Total Registered on VAHAN 4.0']]
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].bar(transport_data['Vehicle Class'], transport_data['Total Registered on VAHAN 4.0'])
ax[0].set_title('Transport Vehicles')
ax[0].set_xlabel('Vehicle Class')
ax[0].set_ylabel('Total Registered')
ax[0].tick_params(axis='x', rotation=90)
ax[1].bar(non_transport_data['Vehicle Class'], non_transport_data['Total Registered on VAHAN 4.0'])
ax[1].set_title('Non-Transport Vehicles')
ax[1].set_xlabel('Vehicle Class')
ax[1].set_ylabel('Total Registered')
ax[1].tick_params(axis='x', rotation=90)

plt.show()
```

```python
import folium
map_center=[20.5937, 78.9629]
map_zoom=5

charging_stations_map=folium.Map(location=map_center, zoom_start=map_zoom)

city_color ='green'
highway_color ='blue'

city_stations =[
    {'location': [30.7333, 76.7794], 'name': 'Chandigarh', 'count': 48},
    {'location': [28.7041, 77.1025], 'name': 'Delhi', 'count': 94},
    {'location': [26.9124, 75.7873], 'name': 'Jaipur', 'count': 49},
    {'location': [12.9716, 77.5946], 'name': 'Bengaluru', 'count': 60},
    {'location': [23.3441, 85.3096], 'name': 'Ranchi', 'count': 30},
    {'location': [26.8467, 80.9462], 'name': 'Lucknow', 'count': 1},
    {'location': [15.2993, 74.1240], 'name': 'Goa', 'count': 30},
    {'location': [17.3850, 78.4867], 'name': 'Hyderabad', 'count': 57},
    {'location': [27.1767, 78.0081], 'name': 'Agra', 'count': 15},
    {'location': [31.1048, 77.1734], 'name': 'Shimla', 'count': 9}
]
for station in city_stations:
    folium.Marker(location=station['location'], popup=f"{station['name']}: {station['count']} charging stations", 
                  icon=folium.Icon(color=city_color)).add_to(charging_stations_map)

highway_stations =[
    {'location': [30.7333, 76.7794], 'name': 'Delhi - Chandigarh', 'count': 24},
    {'location': [18.5204, 73.8567], 'name': 'Mumbai - Pune', 'count': 16},
    {'location': [27.0238, 74.2179], 'name': 'Delhi - Jaipur - Agra', 'count': 31},
    {'location': [26.8853, 75.7910], 'name': 'Jaipur - Delhi Highway', 'count': 9}
]
for station in highway_stations:
    folium.Marker(location=station['location'], popup=f"{station['name']}: {station['count']} charging stations", 
                  icon=folium.Icon(color=highway_color)).add_to(charging_stations_map)

charging_stations_map
```
